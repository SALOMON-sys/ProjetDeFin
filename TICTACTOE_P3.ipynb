```python
# Étape 1 : Initialisation

import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("TIC-TAC-TOE")

chiffres = [1, 2, 3, 4, 5, 6, 7, 8, 9]  
mark = ''                             
count = 0                               
panneaux = ["panneau"] + ["" for _ in range(9)] 


# Étape 2 : Vérification du jeu

# Fonction pour tester si un joueur a gagné
def win(player):
    return ((panneaux[1] == panneaux[2] == panneaux[3] == player) or
            (panneaux[4] == panneaux[5] == panneaux[6] == player) or
            (panneaux[7] == panneaux[8] == panneaux[9] == player) or
            (panneaux[1] == panneaux[4] == panneaux[7] == player) or
            (panneaux[2] == panneaux[5] == panneaux[8] == player) or
            (panneaux[3] == panneaux[6] == panneaux[9] == player) or
            (panneaux[1] == panneaux[5] == panneaux[9] == player) or
            (panneaux[3] == panneaux[5] == panneaux[7] == player))

# Fonction principale de mise à jour après un clic
def check(digit):
    global count, mark, chiffres

    # Vérifier si la case est libre
    if digit in chiffres:
        chiffres.remove(digit) 
        count += 1

        # Déterminer le joueur courant
        if count % 2 == 0:
            mark = 'O'
        else:
            mark = 'X'

        # Mettre à jour le plateau et le bouton
        panneaux[digit] = mark
        boutons[digit-1].config(text=mark, state="disabled")

        # Vérifier victoire
        if win(mark):
            if mark == 'X':
                messagebox.showinfo("Résultat", "Joueur 1 (X) gagne !")
            else:
                messagebox.showinfo("Résultat", "Joueur 2 (O) gagne !")
            root.destroy()

        # Vérifier égalité
        elif count > 8:
            messagebox.showinfo("Résultat", " Match à égalité !")
            root.destroy()

# Partie 3 : Interface graphique

# Étiquettes des joueurs
label_j1 = tk.Label(root, text="Joueur 1 : X", font=('Arial', 14))
label_j1.grid(row=0, column=0, pady=5)

label_j2 = tk.Label(root, text="Joueur 2 : O", font=('Arial', 14))
label_j2.grid(row=0, column=2, pady=5)

# Plateau de jeu (3x3 boutons)
boutons = []
for i in range(1, 10):
    b = tk.Button(root, text=" ", font=('Arial', 20), width=6, height=3,
                  command=lambda i=i: check(i))
    b.grid(row=(i-1)//3 + 1, column=(i-1)%3, padx=5, pady=5)
    boutons.append(b)


# Lancer l'application
root.mainloop()
```
